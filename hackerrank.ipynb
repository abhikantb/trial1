```python
input_arr=[2,4,2,3,1,3,7,5,3] 
min_input,max_input,input_length=min(input_arr),max(input_arr),len(input_arr)
count_arr=[0]*(max_input-min_input+1)
output_arr=[0]*input_length

for i in range(input_length):
    count_arr[input_arr[i]-min_input]+=1     #counting numbers [1, 2, 3, 1, 1, 0, 1]
for i in range(1,len(count_arr)):
    count_arr[i]+=count_arr[i-1]             # cumulative addition [1, 3, 6, 7, 8, 8, 9]
for i in range(input_length-1,-1,-1):
    output_arr[count_arr[input_arr[i]-min_input]-1]=input_arr[i]   #place input array element in output array 
    count_arr[input_arr[i]-min_input]-=1     #decrease count in count arr by 1
output_arr
```

```python
def pangram(sentence1):    
    sentence1=sentence1.lower()
    # print(chr(97+25))
    set_char={chr(i) for i in range(97,97+26)}
    set_char.add(' ')
    count=0
    for i in set_char:
        if i in sentence1:
            pass
        else:
            count+=1
    if count>0:
        return 'not pangram'
    else:
        return 'pangram'
sentence1='We promptly judged antique ivory buckles for the next prize'
pangram(sentence1)
```
